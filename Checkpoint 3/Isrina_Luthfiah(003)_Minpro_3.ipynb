**Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

**COMPARISON** ‧₊ ᵎᵎ 🍒 ⋅ ˚✮

Visualisasi pH Tanah Untuk Setiap Jenis Tanaman 🌱 Untuk memberikan gambaran yang jelas tentang kondisi tanah yang ideal bagi berbagai tanaman.

```python
# Bar Chart (Matplotlib): Kondisi pH Tanah per jenis Tanaman
plt.figure(figsize=(8, 6))
env_conditions = df.groupby("label").mean()
plt.bar(env_conditions.index, env_conditions['ph'], color='olivedrab')
plt.title('Kebutuhan pH Tanah Rata-rata untuk Setiap Jenis Tanaman', fontsize=16)
plt.ylabel('pH Tanah', fontsize=14)
plt.xlabel('Jenis Tanaman', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

📊Insight Bar Chart:

Grafik ini menggambarkan kebutuhan pH tanahh rata-rata dimana setiap jenis tanaman memiliki tingkat pH tanah yang ideal untuk tumbuh dengan baik. Beberapa tanaman, seperti apple dan kidneybeans, lebih cocok di tanah yang cenderung asam (pH di bawah 6). Sementara itu, tanaman seperti banana dan chickpea lebih optimal tumbuh di tanah yang netral hingga sedikit basa (pH antara 6.5 hingga 7.5).

👨‍🌾 Tindakan:

Dengan memahami pola kebutuhan pH tanah untuk setiap jenis tanaman, petani dapat melakukan tindakan yang lebih tepat dalam pemilihan lahan, jenis pupuk, serta metode pengolahan tanah. **COMPOSITION**  ‧₊ ᵎᵎ 🍒 ⋅ ˚✮

Pie chart menyajikan distribusi rata-rata faktor pertanian dengan jelas. Setiap segmen menunjukkan persentase dari unsur hara, suhu, kelembapan, dan pH, sehingga dapat memudahkan untuk identifikasi faktor yang dominan.

```python
mean_values = df[['N', 'P', 'K', 'temperature', 'humidity', 'ph']].mean()
plt.figure(figsize=(8, 8))
plt.pie(mean_values, labels=mean_values.index, autopct='%1.1f%%', startangle=140,
        colors=['lightcoral', 'gold', 'lightgreen', 'lightskyblue', 'plum', 'orange'])
plt.title("Komposisi Rata-rata Faktor Pertanian")
plt.show()
```

📊Insight Histogram:

Jika kelembapan dan suhu mendominasi, artinya kondisi lingkungan sangat mempengaruhi hasil pertanian. Jika kandungan N, P, dan K tinggi, berarti tanah kaya nutrisi, namun perlu penyesuaian agar seimbang melalui pemupukan yang tepat.

👨‍🌾Tindakan:
Kondisi lingkungan yang didominasi oleh suhu dan kelembapan menunjukkan bahwa hasil pertanian sangat dipengaruhi oleh faktor-faktor tersebut, sehingga penting untuk memilih varietas tanaman yang tahan terhadap kondisi ekstrem dan mengatur waktu tanam yang tepat. Sementara itu, meskipun tanah kaya nutrisi jika kandungan N, P, dan K tinggi, penggunaan pupuk seimbang diperlukan agar nutrisi tetap optimal. Pemantauan rutin kondisi tanah dan lingkungan juga penting untuk menyesuaikan intervensi secara tepat waktu.

**RELATIONSHIP** ‧₊ ᵎᵎ 🍒 ⋅ ˚✮

Heatmap digunakan untuk menunjukkan hubungan antara variabel dalam dataset, membantu memahami keterkaitan antara unsur hara dan faktor lingkungan.

```python
# Heatmap (Seaborn): Korelasi Antara Unsur Hara dan Faktor  Lingkungan
plt.figure(figsize=(8, 6))
sns.heatmap(df[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'THI']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Unsur Hara dan Faktor Lingkungan')
plt.show()
```

📊Insight Heatmap:

Dari heatmap korelasi, terlihat hubungan antara unsur hara dan faktor lingkungan. Misalnya, kandungan fosfor (P) dan kalium (K) memiliki korelasi tinggi (+0.74), yang berarti jika P tinggi, K cenderung tinggi juga, ini bisa dimanfaatkan dalam manajemen pemupukan untuk efisiensi biaya.

Suhu (temperature) berkorelasi kuat dengan THI (+0.86), begitu juga kelembaban udara (+0.66), menandakan bahwa peningkatan suhu dan kelembaban berkontribusi terhadap indeks ini. Petani perlu memantau faktor ini agar kondisi tanaman tetap optimal.
